```python
## Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
## Reading the csv files
df1 = pd.read_csv('Total_population_countrywise.csv',delimiter='\t')
df1.head()
```

```python
df1.info()
```

```python
df1.isna().sum()
```

### The first row is empty therefore we should drop it. After dropping the first row, the indices should be reset.

```python
df1 = df1.drop(index=0).reset_index(drop=True)
df1.head()
```

```python
## Setting the first row as column names
df1.columns = df1.iloc[0]
df1 = df1.drop(0)
df1 = df1.reset_index(drop=True)
df1.head()
```

```python
data = df1.set_index('Country Name').iloc[:,3:].T
```

```python
data = data.apply(pd.to_numeric, errors = 'coerce').dropna(how='all')
total_population = data.sum(axis=1)
```

```python
data.head()
```

```python
total_population
```

## Plotting The bar charts

```python
plt.figure(figsize=(15,6))
total_population.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Total Population Over Years')
plt.show()
```
